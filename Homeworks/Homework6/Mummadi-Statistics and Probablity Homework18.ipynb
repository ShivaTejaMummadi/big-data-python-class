
## 6.1.1
Write a function, flip_sum,  which generates $n$ random coin flips from a fair coin and then returns the number of heads. 

A fair coin is defined to be a coin where $P($heads$)=\frac{1}{2}$

The output type should be a numpy integer, hint: use random.rand()


STEPS:
- Assuming that for an random even number its Head and for odd number its tails
- if 2 % random number is 0. Its an even number and its for Heads
- otherwise the number would be 1 and its for tails
- When even number incrementing the count and returning the count at the end.

```python
def flip_sum(flips):
    count = 0
    for i in range(0,flips):
        x = randint(0,100)
        #if even number assuming its a Head and Odd number assuming its a tail
        if x % 2 == 0:
            count += 1
        else:
            continue
    return count
```

## 6.1.4
Write a function, estimate_prob, that uses flip_sum to estimate the following probability:

$P(  k_1 <= $ number of heads in $n$ flips $< k_2 ) $

The function should estimate the probability by running $m$ different trials of flip_sum(n), probably using a for loop.

In order to receive full credit estimate_prob call flip_sum (aka: flip_sum is located inside the estimate_prob function)


STEPS
- pass the limits k1 and k2 into the definition
- and if the headcount for that particular scenario falls in the range then increment 1 else just continue
- At the end return the probability

```python
def estimate_prob(n,k1,k2,m):
    
    count = 0
    for i in range(0, m):
        headCount = flip_sum(n)
        
        if k1 <= headCount and headCount <k2:
            count +=1
           # print 1
        else:
            continue
    #print count, m
    return count/float(m)
```

3. The person is a female, given that the person opposes the issue.

STEPS

- Copy the elements into the list separately for Males and Females
- Calculate the probabilities individually.
- Apply Bayes theorem to calculate the conditional probability

```python
M = [75,89,10]
F = [105,56,6]

persons = sum(M)+sum(F)
#1) No opinion
print "No opinion probability:", (M[2]+F[2])/float(persons)

#2) Male OR against issue
print "Male OR against issue probability:",(sum(M)+F[1])/float(persons)

opp = (M[1]+F[1])/float(persons)
fempro=sum(F)/float(persons)
fem = F[1]/float(persons)
#3) Female, given that person opposes the issue.
print "Female, given that person opposes the issue", (fempro*fem)/float(opp)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

# Load the diabetes dataset
diabetes = datasets.load_diabetes()


# Use only one feature
diabetes_X = diabetes.data[:, np.newaxis, 2]

# Split the data into training/testing sets
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]

# Split the targets into training/testing sets
diabetes_y_train = diabetes.target[:-20]
diabetes_y_test = diabetes.target[-20:]

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(diabetes_X_train, diabetes_y_train)

# Make predictions using the testing set
diabetes_y_pred = regr.predict(diabetes_X_test)

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(diabetes_y_test, diabetes_y_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(diabetes_y_test, diabetes_y_pred))

# Plot outputs
plt.scatter(diabetes_X_test, diabetes_y_test,  color='black')
plt.plot(diabetes_X_test, diabetes_y_pred, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```
