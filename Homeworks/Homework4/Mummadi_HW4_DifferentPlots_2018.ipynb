```python
import numpy as np
import matplotlib.pyplot as plt


def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

# adjusting points with different steps 
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

#adding first subplot 
plt.figure(1)

# 211 means 2 rows 1 column and 1 grid in the figure 
plt.subplot(211)

plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

# 212 means 2 rows 1 column and 2nd grid in the figure
plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```
