Created a Stochastic matrix based on the results of crawling.
At the end created a data frame and replaced unavailable values to 0 and nan_to_num converts a infinite value to largest number  

```python
import pandas as pd
import numpy as np

lst = []
raw = pd.read_csv("steam_txt.csv")
raw['link'] = 'https://www.engadget.com'+raw['link']
r = raw['response'][0]
lst.append((r.split()[-1]).split('>')[-2])
for i in range(len(raw)):
    if raw['response'][i] == r:
        lst.append(raw['link'][i])
    else:
        r = raw['response'][i]
        lst.append((r.split()[-1]).split('>')[-2])
        
item = list(pd.DataFrame(lst)[0].unique())

link = []
length = len(lst)
for i, val in enumerate(lst):
    if i < length-1:
        link.append((lst[i], lst[i+1]))

n = pd.DataFrame(index=item, columns=item)
m = n.replace(np.NaN, 0)
for i in link:
    m.loc[i] = 1.0

ar = np.array(m)
v = ar.sum(axis=1)

result = ar/v[:, np.newaxis]
matrix = np.nan_to_num(result)

print array(matrix)
```

Implemented page rank similar to the lecture notes

```python
from numpy import *

def pagerank(H):
    n = len(H)
    w = zeros(n)
    rho = 1./n * ones(n);
    for i in range(n):
        if multiply.reduce(H[i]== zeros(n)):
            w[i] = 1
    newH = H + outer((1./n * w),ones(n))
 
    theta=0.85
    G = (theta * newH) + ((1-theta) * outer(1./n * ones(n), ones(n)))
    #print rho
    for j in range(10):
        rho = dot(rho,G)
        #print rho
    
    return list(rho)
```
